# Desafio: como o ensino sobre tecnologia pode ser acessível a todos?

Em 2021, o *iFood* assumiu três compromissos públicos na área de educação para serem atingidos em 2025. Essas metas têm o objetivo de formar mais profissionais em tecnologia e prepará-los para o futuro, além de combater o apagão tecnológico no Brasil e no mundo com o objetivo de formar e empregar 25 mil pessoas de
baixa renda em tecnologia.

Os três compromissos públicos estão conectados diretamente com os objetivos de desenvolvimento sustentável (ODS) da ONU. As ODS representam um plano de ação global com 17 metas para eliminar a pobreza extrema e a fome, oferecer educação de qualidade ao longo da vida para todos, proteger o planeta e promover sociedades pacíficas e inclusivas até 2030.

O ensino e a capacitação de brasileiros e brasileiras é importante para a empresa porque um de seus propósitos é alimentar o futuro também com educação e, com isso, contribuir para que o país seja mais próspero e menos desigual.


#Problemas

* Vácuo Inspiracional

> Remodelar a forma de inspirar as pessoas com acesso limitado à informação com novas perspectivas de realização pessoal.

* Apagão Tecnológico

>Ajustar o crescimento da necessidade de soluções digitais pelas empresas com a formação tecnológica de profissionais para suprir essa demanda.

* Acesso à Formação Gratuita e de Qualidade

>Oferecer formas de acesso à parte majoritária da população brasileira aos conteúdos previstos de profissionalização na área de Tecnologia da Informação, se integrando com o programa da linha de fomento *Qualifica Mais Itinerário da Formação Técnica e Profissional* do Ministério da Educação, formalizando parcerias entre as instituições para proporcionar aos estudantes do ensino médio mais oportunidades de cursar o Itinerário da Formação Técnica e Profissional.

* Fomento da Empregabilidade

>Revitalizar as pontes de comunicação entre empresas e mão-de-obra qualificada por sistemas de *newsletter* e divulgação interna entre parceiros.

##9.Principais Impedimentos para Aderência da População em Geral do Aprendizado *Tech*

* Baixa renda: grande parte da população não possue computador em casa. Educação: Existem milhões de analfabetos; 

* Ensino fundamental e médio fraco; base matemática e raciocínio lógico fracos; falta de ajuda financeira pra se dedicar aos estudos;

* Falta de interesse, falta de informação e não entender como funciona;

* Falta de acesso na primeira infância; dificuldade em acessar qualquer equipamento; falta de incentivo;

* Acesso aos recursos tecnológicos; rotina de "desalento" e a própria falta de interesse;

* Falta de recursos tecnológicos; medo de tecnologia; custo elevado;

* Financeiro, tempo e conhecimento;

* Desconhecimento, linguagem difícil, insegurança;

* Desnivelamento social e falta de acesso às tecnologias;

* Preço, dificuldade de acesso e desinteresse de pessoas mais velhas e conservadoras em relação ao ensino;

* Acesso, situação econômica e resistência à mudança das gerações mais velhas;

* Valor, dificuldade de acesso, preconceito com o esteriótipo do profissional da área;

* Acesso a um material de qualidade; acesso aos dispositivos para estudo; incentivo ao estudo;

* Falta de interesse; deficiência ou até mesmo a nula existência de aulas de tecnologia, não sendo considerado algo importante em escolas; falta de profissionais adequados para lecionar nesta área;

* Desinteresse; professores que não ensinam bem; aulas entediantes;

* Acesso a internet; acesso à tecnologia; desconhecimento sobre o conteúdo para ter o interesse pessoal em evoluir na área.

#Proposta de Solução: *Programa Potência Tech Jovem*

Uma plataforma que conecta alunos com dificuldades de acesso digital oriundos de escolas públicas à Área de Tecnologia. A plataforma educacional foi projetada para funcionar de forma gratuita, oferecendo uma trilha de formação completa para jovens do ensino médio darem os primeiros passos na área de tecnologia.

#Mercado-Alvo

* Instituições Públicas de Ensino

>Atores principais para divulgação nas comunidades onde estão inseridos: redes estaduais, distrital ou municipais responsáveis pelo Ensino Médio Público.

* Alunos e Corpo Docente

>Pessoas que desejam fazer parte de uma nova jornada tecnológica e acreditam no poder transformador de se apoderar do conhecimento *tech.*

* Empresas Parceiras

>Organizações públicas e/ou privadas que geram desenvolvimento de impacto através de seus negócios inseridos nos locais em que atuam.
